```python
import pandas as pd
from rdkit import Chem
from rdkit import RDLogger

# 1. Silence RDKit warnings/errors
RDLogger.DisableLog('rdApp.*')

# 1. Load your CSV
df = pd.read_csv('/content/RT/SMRT_dataset.csv', sep=';')  # <-- Replace with your file

# 2. Pre-allocate a smiles list
smiles_list = []

# 3. Fast bulk conversion
for inchi in df['inchi'].fillna(''):
    if inchi.strip() == '':
        smiles_list.append(None)
    else:
        try:
            mol = Chem.MolFromInchi(inchi)
            if mol:
                smiles = Chem.MolToSmiles(mol)
                smiles_list.append(smiles)
            else:
                smiles_list.append(None)
        except Exception as e:
            print(f"⚠️ Error on InChI: {inchi[:30]}... {e}")
            smiles_list.append(None)

# 4. Add smiles list to DataFrame
df['smiles'] = smiles_list

# 5. Keep only 'rt' and 'smiles' columns
df_final = df[['pubchem','rt', 'smiles']]

# (Optional) Drop rows with missing SMILES
# df_final = df_final.dropna(subset=['smiles'])

# 6. Save to a new CSV
df_final.to_csv('output_rt_smiles.csv', index=False)

print("✅ Ultra-fast conversion done! Saved as 'output_rt_smiles.csv'.")
```

```python
import pandas as pd
from rdkit import Chem
from tqdm import tqdm

# 1. Load your CSV
df = pd.read_csv('output_rt_smiles.csv')  # <- your file

# 2. Setup tqdm to show progress
tqdm.pandas()

# 3. Define function to validate SMILES
def is_valid_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        return mol is not None
    except:
        return False

# 4. Apply validation with progress bar
df['is_valid'] = df['smiles'].progress_apply(is_valid_smiles)

# 5. Save a new CSV
df.to_csv('output_rt_smiles_validated.csv', index=False)

print("✅ Done! Validation finished. See 'is_valid' column in 'output_rt_smiles_validated.csv'.")
```

```python
import pandas as pd
from rdkit import Chem
from tqdm import tqdm

# 1. Load the CSV (already validated or original)
df = pd.read_csv('output_rt_smiles_validated.csv')  # <- use your validated file

# 2. Setup tqdm
tqdm.pandas()

# 3. Define function to fix/canonicalize SMILES
def fix_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            # Standardize molecule
            Chem.SanitizeMol(mol)  # This corrects valence, charges, etc.
            fixed_smiles = Chem.MolToSmiles(mol, canonical=True)
            return fixed_smiles
        else:
            return None
    except:
        return None

# 4. Apply fixing only on valid SMILES
df['fixed_smiles'] = df['smiles'].where(df['is_valid']).progress_apply(fix_smiles)

# 5. Save the cleaned file
df[['rt', 'fixed_smiles']].dropna().to_csv('output_rt_fixed_smiles.csv', index=False)

print("✅ Done! Fixed and canonical SMILES saved in 'output_rt_fixed_smiles.csv'.")
```
